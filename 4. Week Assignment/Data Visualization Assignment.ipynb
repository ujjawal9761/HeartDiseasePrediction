## Q.2 
- Plot a Sine Wave with color code with "black" color and style as dashdot, also mark "x" and "y" label and give a suitable title to it

```python
x=np.linspace(0,10,1000)
```

```python
# plotting a sine wave curve for the give no in x
plt.plot(x,np.sin(x),color='black',linestyle='dashdot')

# defining the label of x axis
plt.xlabel("Number",size=20,color='yellow')

# defining the label of y axis
plt.ylabel("Sine value",size=20,color='red')

# defining the title of the graph
plt.title("Sine wave curve",size=20)
```

## Q.4

-Plot Stacked Bar Graph for the above data.
    - Use proper colors and x and y label for the graph
    - Use legend positioned at upper right corner
    - Also mark xticks with an angle of 90 Degree

```python
# list having all the cities name
store=['Delhi','Mumbai','Chennai','Kolkata','Hyderabad']
# array contaning clothing data for all the cities

clothing=np.array([82,78,48,46,39])

# array contaning equipment data for all the cities
equipment=np.array([48,31,51,29,18])

# array contaning accessories data for all the cities
accessories=np.array([15,20,14,13,9])

ind=[x for x , _ in enumerate(store)]

# plotting bar graph for acc as cloth and equi as the bottom
plt.bar(ind,accessories,width=0.8,label='accessories',color='black',bottom=equipment+clothing)

# plotting bar grapg for equi and cloth as the bottom
plt.bar(ind,equipment,width=0.8,label='equipment',color='yellow',bottom=clothing)

# plotting bar graph for clothing as the bottom
plt.bar(ind,clothing,width=0.8,label='clothing',color='red')


# defining xticks with a rotation of 90 degree
plt.xticks(ind,store,rotation=90,size=10)

# defining the label of x axis with text size
plt.xlabel("Cities",size=20)

# defining the label of y axis with proper text size
plt.ylabel("Items",size=20)

# defining the position of the legend acc to the question
plt.legend(loc='upper right')
plt.grid(False)

# function to show the above defined graph
plt.show()
```

## Q.5 For Titanic Dataset
- Load Dataset using Seaborn
- Plot a Bar Graph of Survival of Passenger against the Passenger Class.
- Use proper "hue" attribute, and palette color. Mark 'x' and 'y' label with proper title.
- Use legend located at upper left corner

```python
# loading the preloaded titanic dataset in the seaborn package
data = sns.load_dataset('titanic')
data
```

```python
# plotting a bar graph with proper hue, palette , and proper legend position
sns.barplot(x='class',y='survived',hue='class',palette='Accent_r',data=data)

# defining the position of the legend
plt.legend(loc='upper left')

#defining the label of x axis
plt.xlabel("Passenger_class",size=20,color="red")

#deining the label of y axis
plt.ylabel("Survived",size=20,color="red")

#defining the title of the graph
plt.title("Graph to show passenger survived")
plt.show()
```

## Q.6 For Tips Dataset
- Load the Dataset Using Seaborn
- Plot the Horizontal Distribution Graph on Total Bill with color as Green
- Mark proper 'x' and 'y' label and give a suitable title

```python
# loading the preloaded titanic dataset in the seaborn package
dataset=sns.load_dataset('tips')
dataset
```

```python
# plotting the given Horizontal Distribution Graph 
sns.distplot(dataset.total_bill,color='Green')

# marking x label
plt.xlabel("Total_bill",size=20)

# marking y label
plt.ylabel("Score",size=20)
plt.show()

```

## Q.9 


- Plot the Pie Chart for the two colums with some explode for the country "India"  and proper shadow

```python
# defining a list of countries
countries=['United States','Great Britain','India','Russia','Germany']
# total meddles won per country
medals=[46,27,26,19,17]
# to define color of each portion
colors=['green','red','yellow','black','lightblue']
#to explode the india part
explode=(0,0,0.2,0,0)
# function to plot pie chart with given conditions and specification
plt.pie(medals,explode=explode,labels=countries,colors=colors,shadow=True,startangle=200,autopct='%1.1f%%')
plt.show()
```
